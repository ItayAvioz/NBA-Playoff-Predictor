```python
import pandas as pd
import os

# Directory where the CSV files are located
directory = r'C:\Users\yonatanam\Desktop\NBA\Data'

# List all CSV files in the directory
csv_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('_Opp.csv')]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each CSV file
for file in csv_files:
    # Extract year from the file name
    year = os.path.basename(file).split('_')[0]
    
    # Read CSV file
    df = pd.read_csv(file)
    
    # Add a new column 'Year' with the extracted year
    df['Year'] = year
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into one
opponent_avg_by_year_data = pd.concat(dfs, ignore_index=True)

#the combined DataFrame

opponent_avg_by_year_data
```

```python
opponent_avg_by_year_data.to_csv(os.path.join(directory, 'opponent_avg_by_year_data.csv'), index=False)
```
