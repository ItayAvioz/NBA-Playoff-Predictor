# Loading the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.metrics as metrics 
```

```python
# NBA_EDA - 2
X_train = pd.read_csv('C:/Users/yonatanam\Desktop/NBA/X_train_2.1B2.csv')
y_train = pd.read_csv('C:/Users/yonatanam\Desktop/NBA/y_train_2.1B2.csv')

X_dev = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/X_dev_2.1B2.csv')
y_dev = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/y_dev_2.1B2.csv')

X_test = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/X_test_2.1B2.csv')
y_test= pd.read_csv('C:/Users/yonatanam/Desktop/NBA/y_test_2.1B2.csv')
```

```python
def classificationMetrics(y, yhat, threshold=0.5):
    # Convert continuous predictions to binary labels based on threshold
    y_pred_binary = (yhat >= threshold).astype(int)
    
    # Compute accuracy score
    accuracy = metrics.accuracy_score(y, y_pred_binary)
    
    return  accuracy
```

# Train the models

```python
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# Initialize models
models = {
    'Logistic Regression': SGDRegressor(),
    'Decision Trees': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(probability=True),
    'AdaBoost': AdaBoostClassifier()
}

# Train models
for name, model in models.items():
    model.fit(X_train, y_train)
```

# Evaluate models, store results, and plot y_true vs y_pred

```python
results = []
n_features = X_train.shape[1]

for name, model in models.items():
    # Predictions
    y_pred_train = model.predict(X_train)
    y_pred_dev = model.predict(X_dev)
    
    
    # Accuracy for train set
    accuracy_train = classificationMetrics(y_train, y_pred_train)
    
    # Accuracy for dev set
    accuracy_dev = classificationMetrics(y_dev, y_pred_dev)

    # Calculate the weighted averages
    accuracy_50 = (accuracy_dev + accuracy_train) / 2
    accuracy_60 = (0.4 * accuracy_train) + (0.6 * accuracy_dev)
    
    # Append results
    results.append((name, accuracy_train, accuracy_dev, accuracy_50, accuracy_60))
    
    # Plot distribution of predicted probabilities for non-logistic regression models
    if hasattr(model, 'predict_proba'):
        pred_proba_train = model.predict_proba(X_train)[:, 1]
        pred_proba_dev = model.predict_proba(X_dev)[:, 1]
        
        
        # Plot distribution of predicted probabilities for train set
        plt.figure(figsize=(8, 6))
        sns.distplot(pred_proba_train, label='Train Set', color='orange', kde=True)
        plt.title(f'Distribution of Predicted Probabilities (Train Set) - {name}')
        plt.xlabel('Predicted Probability')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Plot distribution of predicted probabilities for dev set
        plt.figure(figsize=(8, 6))
        sns.distplot(pred_proba_dev, label='Dev Set', color='blue', kde=True)
        plt.title(f'Distribution of Predicted Probabilities (Dev Set) - {name}')
        plt.xlabel('Predicted Probability')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Compute crosstabulation for dev set
        crosstab_dev = pd.crosstab(y_dev.values.ravel(), y_pred_dev, rownames=['True Label'], colnames=['Predicted Label'])
        print(f"Crosstabulation for Dev Set - {name}:")
        print(crosstab_dev)
    
        # Compute crosstabulation for train set
        crosstab_train = pd.crosstab(y_train.values.ravel(), y_pred_train, rownames=['True Label'], colnames=['Predicted Label'])
        print(f"Crosstabulation for Train Set - {name}:")
        print(crosstab_train)
    
    # Plot distribution of predicted values for logistic regression models
    if isinstance(model, SGDRegressor):
        # Plot distribution of predicted values for train set
        plt.figure(figsize=(8, 6))
        sns.distplot(y_pred_train, label='Train Set', color='orange', kde=True)
        plt.title(f'Distribution of Predicted Values (Train Set) - {name}')
        plt.xlabel('Predicted Value')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Plot distribution of predicted values for dev set
        plt.figure(figsize=(8, 6))
        sns.distplot(y_pred_dev, label='Dev Set', color='blue', kde=True)
        plt.title(f'Distribution of Predicted Values (Dev Set) - {name}')
        plt.xlabel('Predicted Value')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
```

# Creating dataframe of models comparison

```python
df_results = pd.DataFrame(results, columns=['Model','accuracy_train','accuracy_dev', 'accuracy_50', 'accuracy_60'])
df_results
```
