# Loading the data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.metrics as metrics 
```

```python
# NBA_EDA - 2
X_train = pd.read_csv('C:/Users/yonatanam\Desktop/NBA/X_train_2.1B.csv')
y_train = pd.read_csv('C:/Users/yonatanam\Desktop/NBA/y_train_2.1B.csv')

X_dev = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/X_dev_2.1B.csv')
y_dev = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/y_dev_2.1B.csv')

X_test = pd.read_csv('C:/Users/yonatanam/Desktop/NBA/X_test_2.1B.csv')
y_test= pd.read_csv('C:/Users/yonatanam/Desktop/NBA/y_test_2.1B.csv')
```

```python
#X_train = X_train.drop('noise_2', axis=1)
#X_dev = X_dev.drop('noise_2', axis=1)
#X_test = X_test.drop('noise_2', axis=1)
```

```python
def classificationMetrics(y, yhat, threshold=0.5):
    # Convert continuous predictions to binary labels based on threshold
    y_pred_binary = (yhat >= threshold).astype(int)
    
    # Compute accuracy score
    accuracy = metrics.accuracy_score(y, y_pred_binary)
    
    return  accuracy
```

# Train the models

```python
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

# Initialize models
models = {
    'Logistic Regression': SGDRegressor(),
    'Decision Trees': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(probability=True),
    'AdaBoost': AdaBoostClassifier()
}

# Train models
for name, model in models.items():
    model.fit(X_train, y_train)
```

# Evaluate models, store results, and plot y_true vs y_pred

```python
results = []
n_features = X_train.shape[1]

for name, model in models.items():
    # Predictions
    y_pred_train = model.predict(X_train)
    y_pred_dev = model.predict(X_dev)
    
    
    # Accuracy for train set
    accuracy_train = classificationMetrics(y_train, y_pred_train)
    
    # Accuracy for dev set
    accuracy_dev = classificationMetrics(y_dev, y_pred_dev)

    # Calculate the weighted averages
    accuracy_50 = (accuracy_dev + accuracy_train) / 2
    accuracy_60 = (0.4 * accuracy_train) + (0.6 * accuracy_dev)
    
    # Append results
    results.append((name, accuracy_train, accuracy_dev, accuracy_50, accuracy_60))
    
    # Plot distribution of predicted probabilities for non-logistic regression models
    if hasattr(model, 'predict_proba'):
        pred_proba_train = model.predict_proba(X_train)[:, 1]
        pred_proba_dev = model.predict_proba(X_dev)[:, 1]
        
        
        # Plot distribution of predicted probabilities for train set
        plt.figure(figsize=(8, 6))
        sns.distplot(pred_proba_train, label='Train Set', color='orange', kde=True)
        plt.title(f'Distribution of Predicted Probabilities (Train Set) - {name}')
        plt.xlabel('Predicted Probability')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Plot distribution of predicted probabilities for dev set
        plt.figure(figsize=(8, 6))
        sns.distplot(pred_proba_dev, label='Dev Set', color='blue', kde=True)
        plt.title(f'Distribution of Predicted Probabilities (Dev Set) - {name}')
        plt.xlabel('Predicted Probability')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Compute crosstabulation for dev set
        crosstab_dev = pd.crosstab(y_dev.values.ravel(), y_pred_dev, rownames=['True Label'], colnames=['Predicted Label'])
        print(f"Crosstabulation for Dev Set - {name}:")
        print(crosstab_dev)
    
        # Compute crosstabulation for train set
        crosstab_train = pd.crosstab(y_train.values.ravel(), y_pred_train, rownames=['True Label'], colnames=['Predicted Label'])
        print(f"Crosstabulation for Train Set - {name}:")
        print(crosstab_train)
    
    # Plot distribution of predicted values for logistic regression models
    if isinstance(model, SGDRegressor):
        # Plot distribution of predicted values for train set
        plt.figure(figsize=(8, 6))
        sns.distplot(y_pred_train, label='Train Set', color='orange', kde=True)
        plt.title(f'Distribution of Predicted Values (Train Set) - {name}')
        plt.xlabel('Predicted Value')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
        # Plot distribution of predicted values for dev set
        plt.figure(figsize=(8, 6))
        sns.distplot(y_pred_dev, label='Dev Set', color='blue', kde=True)
        plt.title(f'Distribution of Predicted Values (Dev Set) - {name}')
        plt.xlabel('Predicted Value')
        plt.ylabel('Density')
        plt.xlim(0, 1)  # Set x-axis range from 0 to 1
        plt.legend()
        plt.show()
        
```

```python
# Create an empty list to store the desired rows
desired_rows = []

for i in range(len(data_to_analyze)-1):
    # Get the current row
    row_i = data_to_analyze.iloc[i]

    # Compare the current row with all other rows
    for j in range(i+1, len(data_to_analyze)):
        # Get the next row
        row_j = data_to_analyze.iloc[j]

        # Check conditions:
        # 1. TEAM_ABBREVIATION of row_i equals MATCHUP_TEAM of row_j, and vice versa
        # 2. year, stage, and game_series are equal between row_i and row_j
        if (row_i['TEAM_ABBREVIATION'] == row_j['MATCHUP_TEAM']) and \
           (row_i['MATCHUP_TEAM'] == row_j['TEAM_ABBREVIATION']) and \
           (row_i['year'] == row_j['year']) and \
           (row_i['stage'] == row_j['stage']) and \
           (row_i['game_series'] == row_j['game_series']):
            # Append both rows to the list of desired rows
            desired_rows.append(row_i)
            desired_rows.append(row_j)

# Create a new DataFrame from the list of desired rows
result_df = pd.DataFrame(desired_rows)

result_df
```

```python
# Initialize a new column 'sum_diff' with zeros
result_df['sum_diff'] = 0

# Create a list of row indices
row_indices = list(result_df.index)

# Iterate over the row indices, skipping every other row
for i in range(0, len(row_indices)-1, 2):
    # Get the current and next row indices
    current_row = row_indices[i]
    next_row = row_indices[i+1]

    # Calculate the sum of differences between consecutive rows
    result_df.at[current_row, 'sum_diff'] = abs(result_df.at[current_row, 'y_diff']) + abs(result_df.at[next_row, 'y_diff'])

    # Assign the sum_diff value to the next row as well
    result_df.at[next_row, 'sum_diff'] = result_df.at[current_row, 'sum_diff']

# If there's an odd number of rows, calculate the sum_diff for the last row
if len(result_df) % 2 != 0:
    last_row = row_indices[-1]
    result_df.at[last_row, 'sum_diff'] = result_df.at[last_row, 'y_diff']

result_df
```

# Calibration

```python
# Select only the 'Prediction' column from result_df
prediction_col = result_df[['Prediction']]

# Merge prediction_col with X_train, keeping only rows from prediction_col (result_df)
new_df = prediction_col.merge(X_train, left_index=True, right_index=True, how='left')
new_df
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.calibration import calibration_curve

# Create X and y from your data
X = new_df.drop('Prediction', axis=1)
y = new_df['Prediction']

# Train a random forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100,random_state=143)
rf_classifier.fit(X, y)

# Predict probabilities before calibration
probs_before_calibration = rf_classifier.predict_proba(X)[:, 1]

# Calibrate the classifier using CalibratedClassifierCV with sigmoid calibration
calibrated_classifier = CalibratedClassifierCV(rf_classifier, method='sigmoid')
calibrated_classifier.fit(X, y)

# Predict probabilities after calibration
probs_after_calibration = calibrated_classifier.predict_proba(X)[:, 1]

# Create a new DataFrame with the probabilities
probs_df = pd.DataFrame({
    'Prediction': y,
    'Probability_Before_Calibration': probs_before_calibration,
    'Probability_After_Calibration': probs_after_calibration
})

# Compute calibration curve
prob_true, prob_pred = calibration_curve(y, probs_after_calibration, n_bins=10)

# Plot calibration curve
plt.figure(figsize=(8, 6))
plt.plot(prob_pred, prob_true, marker='o', linestyle='-', color='b', label='Calibration Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Perfectly Calibrated')
plt.xlabel('Mean Predicted Probability')
plt.ylabel('Fraction of Positives')
plt.title('Calibration Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot probabilities before and after calibration
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(probs_df.index, probs_df['Probability_Before_Calibration'], label='Probability Before Calibration', alpha=0.5)
ax.scatter(probs_df.index, probs_df['Probability_After_Calibration'], label='Probability After Calibration', alpha=0.5)
ax.set_xlabel('Sample Index')
ax.set_ylabel('Probability')
ax.set_title('Probabilities Before and After Calibration')
ax.legend()
plt.show()
```
